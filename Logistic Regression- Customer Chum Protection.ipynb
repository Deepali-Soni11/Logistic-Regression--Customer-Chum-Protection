```python
binary_category=['gender','Partner','Dependents','PhoneService','PaperlessBilling','Churn']
multiple_category=['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
       'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
       'Contract','PaymentMethod' ]
```

4. Data Processing

```python
from sklearn.preprocessing import LabelEncoder
le_encoder=LabelEncoder()
for col in binary_category:
    customer_data[col]=le_encoder.fit_transform(customer_data[col])
```

```python
customer_data['PaymentMethod'].unique()
```

```python
customer_df=pd.get_dummies(customer_data, columns=multiple_category,drop_first=True)
```

5. Is data labels imbalance?

```python
customer_df['Churn'].value_counts()
```

```python
sns.catplot(customer_df['Churn'].value_counts(),kind='bar',palette='bwr')
```

```python
X=customer_df.drop(['customerID','Churn'],axis=1)
Y=customer_df['Churn']
```

```python
from imblearn.over_sampling import SMOTE
smote_sample=SMOTE()
transform_feature, transform_label=smote_sample.fit_resample(X,Y)
```

```python
transform_label.value_counts()         #Number of samples: 5163
```

6. Cross Validation
      * Random data split into train & test
 

```python
cols=list(transform_feature.select_dtypes('bool').columns)
```

```python
cols
```

```python
transform_feature[cols]=transform_feature[cols].astype('int')
transform_label=transform_label.astype('int')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(transform_feature,transform_label,test_size=0.2,random_state=0)
```

6. logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#max_iter=maximum iteration, solver='liblinear'(binary classification), penalty=l1(regularisation) & l2(regularisation)
logit_model=LogisticRegression(max_iter=10000,solver='liblinear',penalty='l2')
logit_model.fit(x_train,y_train)
```

```python
#Accuracy
logit_model.score(x_test,y_test)
```

7. Classification Metrics
   * Confusion Matrix
   * Precision, Recall, F1-Score, Sensitivity, Speficity, ROC, AUC etc...

```python
#predicted value
yhat=logit_model.predict(x_test)
```

```python
#predicted probability
logit_model.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test,yhat)
```

```python
sns.heatmap(confusion_matrix(y_test,yhat), annot=True,fmt='0.0f',cmap='Set2')
```

```python
print(classification_report(y_test,yhat))
```
